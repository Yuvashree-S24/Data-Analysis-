```python
questions=["1. How many people have heart disease and how many people doesn't have heart disease?",
           "2. People of which sex has most heart disease?",
           "3. People of which sex which type of chest pain most?"
           "4. People with which chest pain are most pron to have heart disease?",
           "5. What is the correlation between cholesterol and max heart rate in heart disease patients?",
            "6.What percentage of people with high blood pressure (above 140) have heart disease?",
              "7. What is the average age of people with heart disease?"]
print(questions)
```

```python
# 5. What is the correlation between cholesterol and max heart rate in heart disease patients?
correlation = df[df['target'] == 1]['chol'].corr(df[df['target'] == 1]['thalach'])
print("\n5. Correlation between cholesterol and max heart rate in heart disease patients:", correlation)
```

```python
plt.figure(figsize=(6, 6))
sns.scatterplot(data=df[df['target'] == 1], x='chol', y='thalach', color='green')
plt.title(f'Cholesterol vs Max Heart Rate (Correlation = {correlation:.2f})')
plt.xlabel('Cholesterol')
plt.ylabel('Max Heart Rate')
plt.show()
```

```python
# Question 6: Percentage of people with high blood pressure (above 140) having heart disease
high_bp = df[df['trestbps'] > 140]
high_bp_with_heart_disease = high_bp[high_bp['target'] == 1]
percentage_high_bp = (len(high_bp_with_heart_disease) / len(high_bp)) * 100
```

```python
# Visualization for Question 6
labels = ['Heart Disease', 'No Heart Disease']
values = [len(high_bp_with_heart_disease), len(high_bp) - len(high_bp_with_heart_disease)]

plt.figure(figsize=(6, 6))
plt.pie(values, labels=labels, autopct='%1.1f%%', colors=['red', 'green'])
plt.title('High Blood Pressure and Heart Disease')
plt.show()
```

```python
# Question 7: Average age of people with heart disease
avg_age = df[df['target'] == 1]['age'].mean()
```

```python
# Visualization for Question 7
plt.figure(figsize=(6, 4))
sns.histplot(df[df['target'] == 1]['age'], bins=10, kde=True, color='skyblue')
plt.axvline(avg_age, color='red', linestyle='--', label=f'Average Age = {avg_age:.2f}')
plt.title('Age Distribution of Heart Disease Patients')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()
plt.show()
```
